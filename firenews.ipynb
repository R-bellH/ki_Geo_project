```python
import requests

def fetch_fire_news():
    url = "https://api2.effis.emergency.copernicus.eu/firenews/rest/firenews/firenews"
    params = {
        "notify": 1,
        #"place__icontains": "fontana",  # Replace with the required location information
        "startdate__gte": "2024-05-13T00:00:00.000Z",
        "enddate__lte": "2024-05-28T23:59:59.999Z",
        "simpleplace__icontains": "IT",  # 替换为所需的国家/地区的2个字母ISO代码
        "ordering": "-enddate,-startdate",
        "limit": 20,
        "offset": 0
    }

    try:
        response = requests.get(url, params=params)
        response.raise_for_status()  # Check the response status code, if it is not 200, throw an exception
        data = response.json()  # Parse the response content into JSON format
        return data
    except requests.exceptions.RequestException as e:
        print("Error fetching fire news:", e)
        return None


fire_news = fetch_fire_news()
if fire_news:
    print("Fetched fire news successfully:")
    print(fire_news)
else:
    print("Failed to fetch fire news.")
```
